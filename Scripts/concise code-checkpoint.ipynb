```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from scipy.stats import norm
from scipy.special import expit

# ======================
# 1. Prepare Data
# ======================
data = pd.DataFrame({
    'Study': ['Yamamoto et al. 2022', 'Acampa et al. 2020', 'Giubbini et al. 2019',
              'Agostoni et al. 2018', 'Wells et al. 2017', 'Nkoulou et al. 2016'],
    'TP': [13, 12, 26, 5, 8, 7],
    'FP': [2, 3, 49, 0, 0, 4],
    'FN': [7, 2, 6, 1, 1, 3],
    'TN': [21, 8, 81, 24, 20, 14]
})

# Add 0.5 to zero cells to avoid computational issues (Haldane-Anscombe correction)
data[['TP', 'FP', 'FN', 'TN']] = data[['TP', 'FP', 'FN', 'TN']] + 0.5

# ======================
# 2. Define Functions for Meta-Analysis
# ======================
def modifiedHA_meta_analysis(events, totals):
    """Modified Hauck-Anderson random-effects meta-analysis"""
    # Calculate proportions and variances
    props = events / totals
    vars = props * (1 - props) / totals
    
    # Calculate weights (inverse variance)
    weights = 1 / vars
    
    # Calculate pooled estimate (fixed effects)
    pooled = np.sum(weights * props) / np.sum(weights)
    
    # Calculate Cochran's Q and I²
    Q = np.sum(weights * (props - pooled)**2)
    df = len(props) - 1
    I2 = max(0, (Q - df) / Q * 100) if Q > 0 else 0
    
    # Calculate tau² (between-study variance)
    C = np.sum(weights) - np.sum(weights**2) / np.sum(weights)
    tau2 = max(0, (Q - df) / C) if C > 0 else 0
    
    # Calculate random-effects weights
    weights_re = 1 / (vars + tau2)
    
    # Calculate random-effects pooled estimate
    pooled_re = np.sum(weights_re * props) / np.sum(weights_re)
    se_re = np.sqrt(1 / np.sum(weights_re))
    
    # Calculate 95% CI
    ci_low = pooled_re - 1.96 * se_re
    ci_high = pooled_re + 1.96 * se_re
    
    # Calculate p-value
    z = pooled_re / se_re
    p_value = 2 * (1 - norm.cdf(abs(z)))
    
    return {
        'estimate': pooled_re,
        'se': se_re,
        'ci_low': ci_low,
        'ci_high': ci_high,
        'tau2': tau2,
        'Q': Q,
        'I2': I2,
        'p_value': p_value
    }

def calculate_diagnostic_metrics(data):
    """Calculate all diagnostic metrics with meta-analysis"""
    # Sensitivity
    sens = modifiedHA_meta_analysis(data['TP'], data['TP'] + data['FN'])
    
    # Specificity
    spec = modifiedHA_meta_analysis(data['TN'], data['TN'] + data['FP'])
    
    # Positive Likelihood Ratio (PLR)
    plr = sens['estimate'] / (1 - spec['estimate'])
    plr_se = np.sqrt((sens['se']/sens['estimate'])**2 + (spec['se']/(1-spec['estimate']))**2)
    plr_ci_low = np.exp(np.log(plr) - 1.96 * plr_se)
    plr_ci_high = np.exp(np.log(plr) + 1.96 * plr_se)
    
    # Negative Likelihood Ratio (NLR)
    nlr = (1 - sens['estimate']) / spec['estimate']
    nlr_se = np.sqrt((sens['se']/(1-sens['estimate']))**2 + (spec['se']/spec['estimate'])**2)
    nlr_ci_low = np.exp(np.log(nlr) - 1.96 * nlr_se)
    nlr_ci_high = np.exp(np.log(nlr) + 1.96 * nlr_se)
    
    # Diagnostic Odds Ratio (DOR)
    dor = (sens['estimate'] * spec['estimate']) / ((1 - sens['estimate']) * (1 - spec['estimate']))
    dor_se = np.sqrt(sens['se']**2/(sens['estimate']*(1-sens['estimate']))**2 + 
                     spec['se']**2/(spec['estimate']*(1-spec['estimate']))**2)
    dor_ci_low = np.exp(np.log(dor) - 1.96 * dor_se)
    dor_ci_high = np.exp(np.log(dor) + 1.96 * dor_se)
    
    return {
        'Sensitivity': sens,
        'Specificity': spec,
        'PLR': {
            'estimate': plr,
            'ci_low': plr_ci_low,
            'ci_high': plr_ci_high,
            'p_value': 2 * (1 - norm.cdf(abs(np.log(plr)/plr_se)))
        },
        'NLR': {
            'estimate': nlr,
            'ci_low': nlr_ci_low,
            'ci_high': nlr_ci_high,
            'p_value': 2 * (1 - norm.cdf(abs(np.log(nlr)/nlr_se)))
        },
        'DOR': {
            'estimate': dor,
            'ci_low': dor_ci_low,
            'ci_high': dor_ci_high,
            'p_value': 2 * (1 - norm.cdf(abs(np.log(dor)/dor_se)))
        }
    }

# ======================
# 3. Perform Meta-Analysis
# ======================
results = calculate_diagnostic_metrics(data)

# ======================
# 4. Display Results
# ======================
def print_results(metric, results):
    print(f"\n{metric}:")
    print(f"  Estimate: {results[metric]['estimate']:.3f}")
    print(f"  95% CI: [{results[metric]['ci_low']:.3f}, {results[metric]['ci_high']:.3f}]")
    print(f"  p-value: {results[metric]['p_value']:.4f}")
    if 'Q' in results[metric]:
        print(f"  Cochrane Q: {results[metric]['Q']:.2f} (p={1 - norm.cdf(results[metric]['Q'], len(data)-1):.3f})")
        print(f"  I²: {results[metric]['I2']:.1f}%")

print("=== Diagnostic Test Accuracy Meta-Analysis (Random Effects) ===")
print_results('Sensitivity', results)
print_results('Specificity', results)
print_results('PLR', results)
print_results('NLR', results)
print_results('DOR', results)

# ======================
# 5. Additional Cochrane Q Tests
# ======================
def cochran_q_test(events, totals):
    """Calculate Cochrane Q test for homogeneity"""
    props = events / totals
    weights = totals / (props * (1 - props))
    pooled = np.sum(weights * props) / np.sum(weights)
    Q = np.sum(weights * (props - pooled)**2)
    df = len(props) - 1
    p_value = 1 - norm.cdf(Q, df)
    return Q, p_value

print("\n=== Additional Cochrane Q Tests ===")
Q_sens, p_sens = cochran_q_test(data['TP'], data['TP'] + data['FN'])
print(f"Sensitivity Q: {Q_sens:.2f}, p-value: {p_sens:.4f}")

Q_spec, p_spec = cochran_q_test(data['TN'], data['TN'] + data['FP'])
print(f"Specificity Q: {Q_spec:.2f}, p-value: {p_spec:.4f}")
```

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
from scipy.special import expit

# ======================
# 1. Prepare Data (with modified Haldane-Anscombe correction)
# ======================
data = pd.DataFrame({
    'Study': ['Yamamoto et al. 2022', 'Acampa et al. 2020', 'Giubbini et al. 2019',
              'Agostoni et al. 2018', 'Wells et al. 2017', 'Nkoulou et al. 2016'],
    'TP': [13, 12, 26, 5, 8, 7],
    'FP': [2, 3, 49, 0, 0, 4],
    'FN': [7, 2, 6, 1, 1, 3],
    'TN': [21, 8, 81, 24, 20, 14]
})

# Apply modified Haldane-Anscombe correction (only add 0.5 to zero cells)
corrected_data = data.copy()
for col in ['TP', 'FP', 'FN', 'TN']:
    corrected_data[col] = data[col].apply(lambda x: x + 0.5 if x == 0 else x)

# ======================
# 2. Calculate Study-Level Estimates
# ======================
def calculate_study_metrics(df):
    df['Sensitivity'] = df['TP'] / (df['TP'] + df['FN'])
    df['Specificity'] = df['TN'] / (df['TN'] + df['FP'])
    df['PLR'] = df['Sensitivity'] / (1 - df['Specificity'])
    df['NLR'] = (1 - df['Sensitivity']) / df['Specificity']
    df['DOR'] = (df['TP'] * df['TN']) / (df['FP'] * df['FN'])
    
    # Calculate standard errors on logit scale for sensitivity/specificity
    df['logit_Sens'] = np.log(df['Sensitivity'] / (1 - df['Sensitivity']))
    df['logit_Spec'] = np.log(df['Specificity'] / (1 - df['Specificity']))
    df['logit_Sens_SE'] = np.sqrt(1/df['TP'] + 1/df['FN'])
    df['logit_Spec_SE'] = np.sqrt(1/df['TN'] + 1/df['FP'])
    
    # Calculate standard errors for DOR (on log scale)
    df['log_DOR'] = np.log(df['DOR'])
    df['log_DOR_SE'] = np.sqrt(1/df['TP'] + 1/df['FP'] + 1/df['FN'] + 1/df['TN'])
    
    return df

study_metrics = calculate_study_metrics(corrected_data)

# ======================
# 3. Random-Effects Meta-Analysis (DerSimonian-Laird)
# ======================
def random_effects_meta(estimates, standard_errors):
    # Calculate weights (inverse variance)
    weights = 1 / (standard_errors**2)
    
    # Fixed-effects estimate
    pooled = np.sum(weights * estimates) / np.sum(weights)
    
    # Calculate tau-squared (DerSimonian-Laird estimator)
    Q = np.sum(weights * (estimates - pooled)**2)
    df = len(estimates) - 1
    C = np.sum(weights) - np.sum(weights**2) / np.sum(weights)
    tau2 = max(0, (Q - df) / C)
    
    # Recalculate weights with tau2
    new_weights = 1 / (standard_errors**2 + tau2)
    pooled = np.sum(new_weights * estimates) / np.sum(new_weights)
    pooled_se = np.sqrt(1 / np.sum(new_weights))
    
    # Calculate 95% CI
    ci_lower = pooled - 1.96 * pooled_se
    ci_upper = pooled + 1.96 * pooled_se
    
    # Calculate p-value
    z = pooled / pooled_se
    p_value = 2 * (1 - stats.norm.cdf(abs(z)))
    
    return {
        'estimate': pooled,
        'se': pooled_se,
        'ci_lower': ci_lower,
        'ci_upper': ci_upper,
        'p_value': p_value,
        'tau2': tau2,
        'Q': Q,
        'Q_p': 1 - stats.chi2.cdf(Q, df),
        'I2': max(0, (Q - df) / Q * 100) if Q > 0 else 0
    }

# Perform meta-analysis for each metric
meta_sens = random_effects_meta(study_metrics['logit_Sens'], study_metrics['logit_Sens_SE'])
meta_spec = random_effects_meta(study_metrics['logit_Spec'], study_metrics['logit_Spec_SE'])
meta_dor = random_effects_meta(study_metrics['log_DOR'], study_metrics['log_DOR_SE'])

# Back-transform logit and log results
def format_result(meta_result, is_logit=True):
    if is_logit:
        estimate = expit(meta_result['estimate'])
        ci_lower = expit(meta_result['ci_lower'])
        ci_upper = expit(meta_result['ci_upper'])
    else:
        estimate = np.exp(meta_result['estimate'])
        ci_lower = np.exp(meta_result['ci_lower'])
        ci_upper = np.exp(meta_result['ci_upper'])
    
    return {
        'Estimate': estimate,
        '95% CI': f"({ci_lower:.3f}, {ci_upper:.3f})",
        'p-value': meta_result['p_value'],
        'Tau²': meta_result['tau2'],
        'Q': meta_result['Q'],
        'Q p-value': meta_result['Q_p'],
        'I²': f"{meta_result['I2']:.1f}%"
    }

# Calculate PLR and NLR from pooled sensitivity/specificity
sens_pooled = expit(meta_sens['estimate'])
spec_pooled = expit(meta_spec['estimate'])
plr = sens_pooled / (1 - spec_pooled)
nlr = (1 - sens_pooled) / spec_pooled

# ======================
# 4. Prepare Results Table
# ======================
results = pd.DataFrame({
    'Metric': ['Sensitivity', 'Specificity', 'PLR', 'NLR', 'DOR'],
    'Estimate': [
        sens_pooled,
        spec_pooled,
        plr,
        nlr,
        np.exp(meta_dor['estimate'])
    ],
    '95% CI': [
        f"({expit(meta_sens['ci_lower']):.3f}, {expit(meta_sens['ci_upper']):.3f})",
        f"({expit(meta_spec['ci_lower']):.3f}, {expit(meta_spec['ci_upper']):.3f})",
        "Calculated from pooled Sens/Spec",
        "Calculated from pooled Sens/Spec",
        f"({np.exp(meta_dor['ci_lower']):.3f}, {np.exp(meta_dor['ci_upper']):.3f})"
    ],
    'p-value': [
        meta_sens['p_value'],
        meta_spec['p_value'],
        "N/A",
        "N/A",
        meta_dor['p_value']
    ],
    'Tau²': [
        meta_sens['tau2'],
        meta_spec['tau2'],
        "N/A",
        "N/A",
        meta_dor['tau2']
    ],
    'Q': [
        meta_sens['Q'],
        meta_spec['Q'],
        "N/A",
        "N/A",
        meta_dor['Q']
    ],
    'Q p-value': [
        meta_sens['Q_p'],
        meta_spec['Q_p'],
        "N/A",
        "N/A",
        meta_dor['Q_p']
    ],
    'I²': [
        f"{meta_sens['I2']:.1f}%",
        f"{meta_spec['I2']:.1f}%",
        "N/A",
        "N/A",
        f"{meta_dor['I2']:.1f}%"
    ]
})

# ======================
# 5. Output Results
# ======================
print("=== Random-Effects Meta-Analysis Results ===")
print(results.to_string(index=False))

print("\n=== Study-Level Estimates ===")
print(study_metrics[['Study', 'Sensitivity', 'Specificity', 'PLR', 'NLR', 'DOR']].to_string(index=False))

print("\n=== Heterogeneity Statistics ===")
print("Sensitivity:")
print(f"  Q = {meta_sens['Q']:.2f}, p = {meta_sens['Q_p']:.3f}, I² = {meta_sens['I2']:.1f}%")
print("\nSpecificity:")
print(f"  Q = {meta_spec['Q']:.2f}, p = {meta_spec['Q_p']:.3f}, I² = {meta_spec['I2']:.1f}%")
print("\nDOR:")
print(f"  Q = {meta_dor['Q']:.2f}, p = {meta_dor['Q_p']:.3f}, I² = {meta_dor['I2']:.1f}%")
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt

# Data as DataFrame
data = pd.DataFrame({
    'Study': ['Yamamoto et al. 2022', 'Acampa et al. 2020', 'Giubbini et al. 2019',
              'Agostoni et al. 2018', 'Wells et al. 2017', 'Nkoulou et al. 2016'],
    'TP': [13, 12, 26, 5.5, 8.5, 7],
    'FP': [2, 3, 49, 0.5, 0.5, 4],
    'FN': [7, 2, 6, 1.5, 1.5, 3],
    'TN': [21, 8, 81, 24.5, 20.5, 14]
})

# Step 1: Calculate metrics
def calc_metrics(row):
    TP, FP, FN, TN = row['TP'], row['FP'], row['FN'], row['TN']
    sens = TP / (TP + FN) if (TP + FN) > 0 else np.nan
    spec = TN / (FP + TN) if (FP + TN) > 0 else np.nan
    plr = sens / (1 - spec) if (1 - spec) > 0 else np.nan
    nlr = (1 - sens) / spec if spec > 0 else np.nan
    dor = (TP * TN) / (FP * FN) if FP > 0 and FN > 0 else np.nan
    return pd.Series({'Sensitivity': sens, 'Specificity': spec, 'PLR': plr, 'NLR': nlr, 'DOR': dor})

metrics = data.apply(calc_metrics, axis=1)
df = pd.concat([data, metrics], axis=1)

# Step 2: Log-transform
df['lnPLR'] = np.log(df['PLR'])
df['lnNLR'] = np.log(df['NLR'])
df['lnDOR'] = np.log(df['DOR'])

# Step 3: Calculate SE for ln(DOR)
df['SE_lnDOR'] = np.sqrt(1/df['TP'] + 1/df['FP'] + 1/df['FN'] + 1/df['TN'])

# Step 4-9: Fixed-effect and random-effects weights for DOR
df['w_fixed'] = 1 / df['SE_lnDOR']**2
weighted_mean_lnDOR = np.sum(df['w_fixed'] * df['lnDOR']) / np.sum(df['w_fixed'])
Q = np.sum(df['w_fixed'] * (df['lnDOR'] - weighted_mean_lnDOR)**2)
S1 = np.sum(df['w_fixed'])
S2 = np.sum(df['w_fixed']**2)
k = df.shape[0]
tau2 = max(0, (Q - (k - 1)) / (S1 - (S2 / S1)))
df['w_random'] = 1 / (df['SE_lnDOR']**2 + tau2)

# Step 10: Pooled DOR with CI
pooled_lnDOR = np.sum(df['w_random'] * df['lnDOR']) / np.sum(df['w_random'])
pooled_DOR = np.exp(pooled_lnDOR)
SE_pooled_lnDOR = 1 / np.sqrt(np.sum(df['w_random']))
DOR_CI_lower = np.exp(pooled_lnDOR - norm.ppf(0.975) * SE_pooled_lnDOR)
DOR_CI_upper = np.exp(pooled_lnDOR + norm.ppf(0.975) * SE_pooled_lnDOR)

# Step 11: I² statistic
I2 = max(0, (Q - (k - 1)) / Q) * 100

# Step 12: Pooled Sensitivity/Specificity with CI
def logit(p):
    p = np.clip(p, 1e-10, 1 - 1e-10)
    return np.log(p / (1 - p))

def inv_logit(x):
    return 1 / (1 + np.exp(-x))

df['SE_sens'] = np.sqrt(df['Sensitivity'] * (1 - df['Sensitivity']) / (df['TP'] + df['FN']))
df['SE_spec'] = np.sqrt(df['Specificity'] * (1 - df['Specificity']) / (df['FP'] + df['TN']))
df['logit_sens'] = df['Sensitivity'].apply(logit)
df['logit_spec'] = df['Specificity'].apply(logit)

# Sensitivity calculations
df['w_sens_fixed'] = 1 / df['SE_sens']**2
Q_sens = np.sum(df['w_sens_fixed'] * (df['logit_sens'] - np.sum(df['w_sens_fixed'] * df['logit_sens']) / np.sum(df['w_sens_fixed']))**2)
S1_sens = np.sum(df['w_sens_fixed'])
S2_sens = np.sum(df['w_sens_fixed']**2)
tau2_sens = max(0, (Q_sens - (k - 1)) / (S1_sens - (S2_sens / S1_sens)))
df['w_sens_random'] = 1 / (df['SE_sens']**2 + tau2_sens)
pooled_logit_sens = np.sum(df['w_sens_random'] * df['logit_sens']) / np.sum(df['w_sens_random'])
pooled_sens = inv_logit(pooled_logit_sens)
SE_pooled_logit_sens = 1 / np.sqrt(np.sum(df['w_sens_random']))
sens_CI_lower = inv_logit(pooled_logit_sens - norm.ppf(0.975) * SE_pooled_logit_sens)
sens_CI_upper = inv_logit(pooled_logit_sens + norm.ppf(0.975) * SE_pooled_logit_sens)

# Specificity calculations
df['w_spec_fixed'] = 1 / df['SE_spec']**2
Q_spec = np.sum(df['w_spec_fixed'] * (df['logit_spec'] - np.sum(df['w_spec_fixed'] * df['logit_spec']) / np.sum(df['w_spec_fixed']))**2)
S1_spec = np.sum(df['w_spec_fixed'])
S2_spec = np.sum(df['w_spec_fixed']**2)
tau2_spec = max(0, (Q_spec - (k - 1)) / (S1_spec - (S2_spec / S1_spec)))
df['w_spec_random'] = 1 / (df['SE_spec']**2 + tau2_spec)
pooled_logit_spec = np.sum(df['w_spec_random'] * df['logit_spec']) / np.sum(df['w_spec_random'])
pooled_spec = inv_logit(pooled_logit_spec)
SE_pooled_logit_spec = 1 / np.sqrt(np.sum(df['w_spec_random']))
spec_CI_lower = inv_logit(pooled_logit_spec - norm.ppf(0.975) * SE_pooled_logit_spec)
spec_CI_upper = inv_logit(pooled_logit_spec + norm.ppf(0.975) * SE_pooled_logit_spec)

# Step 13: Pooled PLR and NLR with CI
pooled_PLR = pooled_sens / (1 - pooled_spec)
pooled_NLR = (1 - pooled_sens) / pooled_spec

# Delta method for PLR CI
var_sens = (pooled_sens * (1 - pooled_sens))**2 * (1 / np.sum(df['w_sens_random']))
var_spec = (pooled_spec * (1 - pooled_spec))**2 * (1 / np.sum(df['w_spec_random']))
var_logPLR = var_sens / (pooled_sens**2) + var_spec / ((1 - pooled_spec)**2)
SE_logPLR = np.sqrt(var_logPLR)
PLR_CI_lower = np.exp(np.log(pooled_PLR) - norm.ppf(0.975) * SE_logPLR)
PLR_CI_upper = np.exp(np.log(pooled_PLR) + norm.ppf(0.975) * SE_logPLR)

# Delta method for NLR CI
var_logNLR = var_sens / ((1 - pooled_sens)**2) + var_spec / (pooled_spec**2)
SE_logNLR = np.sqrt(var_logNLR)
NLR_CI_lower = np.exp(np.log(pooled_NLR) - norm.ppf(0.975) * SE_logNLR)
NLR_CI_upper = np.exp(np.log(pooled_NLR) + norm.ppf(0.975) * SE_logNLR)

# Step 14: Print results with CIs
print("=== Pooled Estimates with 95% Confidence Intervals ===")
print(f"Pooled Sensitivity: {pooled_sens:.3f} (95% CI: {sens_CI_lower:.3f}-{sens_CI_upper:.3f})")
print(f"Pooled Specificity: {pooled_spec:.3f} (95% CI: {spec_CI_lower:.3f}-{spec_CI_upper:.3f})")
print(f"Pooled PLR: {pooled_PLR:.3f} (95% CI: {PLR_CI_lower:.3f}-{PLR_CI_upper:.3f})")
print(f"Pooled NLR: {pooled_NLR:.3f} (95% CI: {NLR_CI_lower:.3f}-{NLR_CI_upper:.3f})")
print(f"Pooled DOR: {pooled_DOR:.3f} (95% CI: {DOR_CI_lower:.3f}-{DOR_CI_upper:.3f})")

# Step 15: Create forest plots
def create_forest_plot(studies, estimates, lower_bounds, upper_bounds, 
                       pooled_estimate, pooled_lower, pooled_upper,
                       title, xlabel, xlim=None):
    plt.figure(figsize=(10, 6))
    y_pos = np.arange(len(studies) + 1)
    
    # Individual studies
    plt.errorbar(estimates, y_pos, xerr=[np.array(estimates) - np.array(lower_bounds), 
                                         np.array(upper_bounds) - np.array(estimates)],
                 fmt='o', color='black', capsize=5)
    
    # Pooled estimate
    plt.scatter([pooled_estimate], [0], color='red', s=100, zorder=3, label='Pooled Estimate')
    plt.hlines(0, pooled_lower, pooled_upper, color='red', linewidth=2)
    
    # Study labels
    plt.yticks(list(y_pos) + [0], list(studies) + ['Pooled Estimate'])
    plt.axhline(y=0.5, color='grey', linestyle='--')
    plt.title(title)
    plt.xlabel(xlabel)
    
    if xlim:
        plt.xlim(xlim)
    plt.tight_layout()
    plt.show()

# Sensitivity forest plot
studies = df['Study'].tolist()
sens_estimates = df['Sensitivity'].tolist()
sens_lower = [max(0, sens - 1.96 * se) for sens, se in zip(df['Sensitivity'], df['SE_sens'])]
sens_upper = [min(1, sens + 1.96 * se) for sens, se in zip(df['Sensitivity'], df['SE_sens'])]
create_forest_plot(studies, sens_estimates, sens_lower, sens_upper,
                   pooled_sens, sens_CI_lower, sens_CI_upper,
                   'Sensitivity Forest Plot', 'Sensitivity', (0, 1))

# Specificity forest plot
spec_estimates = df['Specificity'].tolist()
spec_lower = [max(0, spec - 1.96 * se) for spec, se in zip(df['Specificity'], df['SE_spec'])]
spec_upper = [min(1, spec + 1.96 * se) for spec, se in zip(df['Specificity'], df['SE_spec'])]
create_forest_plot(studies, spec_estimates, spec_lower, spec_upper,
                   pooled_spec, spec_CI_lower, spec_CI_upper,
                   'Specificity Forest Plot', 'Specificity', (0, 1))

# DOR forest plot
dor_estimates = df['DOR'].tolist()
dor_lower = np.exp(df['lnDOR'] - 1.96 * df['SE_lnDOR'])
dor_upper = np.exp(df['lnDOR'] + 1.96 * df['SE_lnDOR'])
create_forest_plot(studies, dor_estimates, dor_lower, dor_upper,
                   pooled_DOR, DOR_CI_lower, DOR_CI_upper,
                   'Diagnostic Odds Ratio Forest Plot', 'DOR', (0, None))
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt

# Data as DataFrame
data = pd.DataFrame({
    'Study': ['Yamamoto et al. 2022', 'Acampa et al. 2020', 'Giubbini et al. 2019',
              'Agostoni et al. 2018', 'Wells et al. 2017', 'Nkoulou et al. 2016'],
    'TP': [13, 12, 26, 5.5, 8.5, 7],
    'FP': [2, 3, 49, 0.5, 0.5, 4],
    'FN': [7, 2, 6, 1.5, 1.5, 3],
    'TN': [21, 8, 81, 24.5, 20.5, 14]
})

# Step 1: Calculate metrics
def calc_metrics(row):
    TP, FP, FN, TN = row['TP'], row['FP'], row['FN'], row['TN']
    sens = TP / (TP + FN) if (TP + FN) > 0 else np.nan
    spec = TN / (FP + TN) if (FP + TN) > 0 else np.nan
    plr = sens / (1 - spec) if (1 - spec) > 0 else np.nan
    nlr = (1 - sens) / spec if spec > 0 else np.nan
    dor = (TP * TN) / (FP * FN) if FP > 0 and FN > 0 else np.nan
    return pd.Series({'Sensitivity': sens, 'Specificity': spec, 'PLR': plr, 'NLR': nlr, 'DOR': dor})

metrics = data.apply(calc_metrics, axis=1)
df = pd.concat([data, metrics], axis=1)

# Step 2: Log-transform
df['lnPLR'] = np.log(df['PLR'])
df['lnNLR'] = np.log(df['NLR'])
df['lnDOR'] = np.log(df['DOR'])

# Step 3: Calculate SE for ln(DOR)
df['SE_lnDOR'] = np.sqrt(1/df['TP'] + 1/df['FP'] + 1/df['FN'] + 1/df['TN'])

# Step 4-9: Fixed-effect and random-effects weights for DOR
df['w_fixed'] = 1 / df['SE_lnDOR']**2
weighted_mean_lnDOR = np.sum(df['w_fixed'] * df['lnDOR']) / np.sum(df['w_fixed'])
Q = np.sum(df['w_fixed'] * (df['lnDOR'] - weighted_mean_lnDOR)**2)
S1 = np.sum(df['w_fixed'])
S2 = np.sum(df['w_fixed']**2)
k = df.shape[0]
tau2 = max(0, (Q - (k - 1)) / (S1 - (S2 / S1)))
df['w_random'] = 1 / (df['SE_lnDOR']**2 + tau2)

# Step 10: Pooled DOR with CI
pooled_lnDOR = np.sum(df['w_random'] * df['lnDOR']) / np.sum(df['w_random'])
pooled_DOR = np.exp(pooled_lnDOR)
SE_pooled_lnDOR = 1 / np.sqrt(np.sum(df['w_random']))
DOR_CI_lower = np.exp(pooled_lnDOR - norm.ppf(0.975) * SE_pooled_lnDOR)
DOR_CI_upper = np.exp(pooled_lnDOR + norm.ppf(0.975) * SE_pooled_lnDOR)

# Step 11: I² statistic
I2 = max(0, (Q - (k - 1)) / Q) * 100

# Step 12: Pooled Sensitivity/Specificity with CI
def logit(p):
    p = np.clip(p, 1e-10, 1 - 1e-10)
    return np.log(p / (1 - p))

def inv_logit(x):
    return 1 / (1 + np.exp(-x))

df['SE_sens'] = np.sqrt(df['Sensitivity'] * (1 - df['Sensitivity']) / (df['TP'] + df['FN']))
df['SE_spec'] = np.sqrt(df['Specificity'] * (1 - df['Specificity']) / (df['FP'] + df['TN']))
df['logit_sens'] = df['Sensitivity'].apply(logit)
df['logit_spec'] = df['Specificity'].apply(logit)

# Sensitivity calculations
df['w_sens_fixed'] = 1 / df['SE_sens']**2
Q_sens = np.sum(df['w_sens_fixed'] * (df['logit_sens'] - np.sum(df['w_sens_fixed'] * df['logit_sens']) / np.sum(df['w_sens_fixed']))**2)
S1_sens = np.sum(df['w_sens_fixed'])
S2_sens = np.sum(df['w_sens_fixed']**2)
tau2_sens = max(0, (Q_sens - (k - 1)) / (S1_sens - (S2_sens / S1_sens)))
df['w_sens_random'] = 1 / (df['SE_sens']**2 + tau2_sens)
pooled_logit_sens = np.sum(df['w_sens_random'] * df['logit_sens']) / np.sum(df['w_sens_random'])
pooled_sens = inv_logit(pooled_logit_sens)
SE_pooled_logit_sens = 1 / np.sqrt(np.sum(df['w_sens_random']))
sens_CI_lower = inv_logit(pooled_logit_sens - norm.ppf(0.975) * SE_pooled_logit_sens)
sens_CI_upper = inv_logit(pooled_logit_sens + norm.ppf(0.975) * SE_pooled_logit_sens)

# Specificity calculations
df['w_spec_fixed'] = 1 / df['SE_spec']**2
Q_spec = np.sum(df['w_spec_fixed'] * (df['logit_spec'] - np.sum(df['w_spec_fixed'] * df['logit_spec']) / np.sum(df['w_spec_fixed']))**2)
S1_spec = np.sum(df['w_spec_fixed'])
S2_spec = np.sum(df['w_spec_fixed']**2)
tau2_spec = max(0, (Q_spec - (k - 1)) / (S1_spec - (S2_spec / S1_spec)))
df['w_spec_random'] = 1 / (df['SE_spec']**2 + tau2_spec)
pooled_logit_spec = np.sum(df['w_spec_random'] * df['logit_spec']) / np.sum(df['w_spec_random'])
pooled_spec = inv_logit(pooled_logit_spec)
SE_pooled_logit_spec = 1 / np.sqrt(np.sum(df['w_spec_random']))
spec_CI_lower = inv_logit(pooled_logit_spec - norm.ppf(0.975) * SE_pooled_logit_spec)
spec_CI_upper = inv_logit(pooled_logit_spec + norm.ppf(0.975) * SE_pooled_logit_spec)

# Step 13: Pooled PLR and NLR with CI
pooled_PLR = pooled_sens / (1 - pooled_spec)
pooled_NLR = (1 - pooled_sens) / pooled_spec

# Delta method for PLR CI
var_sens = (pooled_sens * (1 - pooled_sens))**2 * (1 / np.sum(df['w_sens_random']))
var_spec = (pooled_spec * (1 - pooled_spec))**2 * (1 / np.sum(df['w_spec_random']))
var_logPLR = var_sens / (pooled_sens**2) + var_spec / ((1 - pooled_spec)**2)
SE_logPLR = np.sqrt(var_logPLR)
PLR_CI_lower = np.exp(np.log(pooled_PLR) - norm.ppf(0.975) * SE_logPLR)
PLR_CI_upper = np.exp(np.log(pooled_PLR) + norm.ppf(0.975) * SE_logPLR)

# Delta method for NLR CI
var_logNLR = var_sens / ((1 - pooled_sens)**2) + var_spec / (pooled_spec**2)
SE_logNLR = np.sqrt(var_logNLR)
NLR_CI_lower = np.exp(np.log(pooled_NLR) - norm.ppf(0.975) * SE_logNLR)
NLR_CI_upper = np.exp(np.log(pooled_NLR) + norm.ppf(0.975) * SE_logNLR)

# Step 14: Print results with CIs
print("=== Pooled Estimates with 95% Confidence Intervals ===")
print(f"Pooled Sensitivity: {pooled_sens:.3f} (95% CI: {sens_CI_lower:.3f}-{sens_CI_upper:.3f})")
print(f"Pooled Specificity: {pooled_spec:.3f} (95% CI: {spec_CI_lower:.3f}-{spec_CI_upper:.3f})")
print(f"Pooled PLR: {pooled_PLR:.3f} (95% CI: {PLR_CI_lower:.3f}-{PLR_CI_upper:.3f})")
print(f"Pooled NLR: {pooled_NLR:.3f} (95% CI: {NLR_CI_lower:.3f}-{NLR_CI_upper:.3f})")
print(f"Pooled DOR: {pooled_DOR:.3f} (95% CI: {DOR_CI_lower:.3f}-{DOR_CI_upper:.3f})")

# Calculate confidence intervals for all metrics
df['Sens_CI_lower'] = np.maximum(0, df['Sensitivity'] - 1.96 * df['SE_sens'])
df['Sens_CI_upper'] = np.minimum(1, df['Sensitivity'] + 1.96 * df['SE_sens'])
df['Spec_CI_lower'] = np.maximum(0, df['Specificity'] - 1.96 * df['SE_spec'])
df['Spec_CI_upper'] = np.minimum(1, df['Specificity'] + 1.96 * df['SE_spec'])
df['DOR_CI_lower'] = np.exp(df['lnDOR'] - 1.96 * df['SE_lnDOR'])
df['DOR_CI_upper'] = np.exp(df['lnDOR'] + 1.96 * df['SE_lnDOR'])

# Calculate SE for PLR and NLR for individual studies
df['SE_lnPLR'] = np.sqrt((1 - df['Sensitivity']) / (df['Sensitivity'] * (df['TP'] + df['FN'])) + 
                      df['Specificity'] / ((1 - df['Specificity']) * (df['FP'] + df['TN'])))
df['SE_lnNLR'] = np.sqrt(df['Sensitivity'] / ((1 - df['Sensitivity']) * (df['TP'] + df['FN'])) + 
                      (1 - df['Specificity']) / (df['Specificity'] * (df['FP'] + df['TN'])))

# Calculate CI for PLR and NLR for individual studies
df['PLR_CI_lower'] = np.exp(df['lnPLR'] - 1.96 * df['SE_lnPLR'])
df['PLR_CI_upper'] = np.exp(df['lnPLR'] + 1.96 * df['SE_lnPLR'])
df['NLR_CI_lower'] = np.exp(df['lnNLR'] - 1.96 * df['SE_lnNLR'])
df['NLR_CI_upper'] = np.exp(df['lnNLR'] + 1.96 * df['SE_lnNLR'])

# Create and display metric tables
def create_metric_table(metric, label, pooled_val, ci_lower, ci_upper):
    table = pd.DataFrame({
        'Study': df['Study'],
        f'{label}': df[metric],
        'Lower 95% CI': df[f'{metric}_CI_lower'],
        'Upper 95% CI': df[f'{metric}_CI_upper']
    })
    
    # Add pooled estimate as last row
    pooled_row = pd.DataFrame({
        'Study': ['Pooled Estimate'],
        f'{label}': [pooled_val],
        'Lower 95% CI': [ci_lower],
        'Upper 95% CI': [ci_upper]
    })
    return pd.concat([table, pooled_row], ignore_index=True)

# Generate all tables
sensitivity_table = create_metric_table(
    'Sensitivity', 'Sensitivity', pooled_sens, sens_CI_lower, sens_CI_upper
)
specificity_table = create_metric_table(
    'Specificity', 'Specificity', pooled_spec, spec_CI_lower, spec_CI_upper
)
plr_table = create_metric_table(
    'PLR', 'PLR', pooled_PLR, PLR_CI_lower, PLR_CI_upper
)
nlr_table = create_metric_table(
    'NLR', 'NLR', pooled_NLR, NLR_CI_lower, NLR_CI_upper
)
dor_table = create_metric_table(
    'DOR', 'DOR', pooled_DOR, DOR_CI_lower, DOR_CI_upper
)

# Print all tables
print("\n\n" + "="*70)
print("Sensitivity Results with 95% Confidence Intervals")
print("="*70)
print(sensitivity_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Specificity Results with 95% Confidence Intervals")
print("="*70)
print(specificity_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Positive Likelihood Ratio (PLR) Results")
print("="*70)
print(plr_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Negative Likelihood Ratio (NLR) Results")
print("="*70)
print(nlr_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Diagnostic Odds Ratio (DOR) Results")
print("="*70)
print(dor_table.round(3))
print("\n" + "-"*70)

# Print heterogeneity metrics
print("\n\nHeterogeneity Metrics:")
print(f"I² Statistic: {I2:.1f}%")
print(f"Tau²: {tau2:.4f}")

# Step 15: Create forest plots
def create_forest_plot(studies, estimates, lower_bounds, upper_bounds, 
                       pooled_estimate, pooled_lower, pooled_upper,
                       title, xlabel, xlim=None, log_scale=False):
    plt.figure(figsize=(10, 6))
    n_studies = len(studies)
    y_pos = np.arange(n_studies)  # Positions for individual studies
    pooled_pos = n_studies  # Position for pooled estimate (below all studies)
    
    # Individual studies
    plt.errorbar(estimates, y_pos, 
                 xerr=[np.array(estimates) - np.array(lower_bounds), 
                       np.array(upper_bounds) - np.array(estimates)],
                 fmt='o', color='black', capsize=5, label='Individual Studies')
    
    # Pooled estimate
    plt.scatter(pooled_estimate, pooled_pos, color='red', s=100, zorder=3, 
                label='Pooled Estimate')
    plt.hlines(pooled_pos, pooled_lower, pooled_upper, color='red', linewidth=3)
    
    # Study labels
    all_labels = list(studies) + ['Pooled Estimate']
    plt.yticks(np.arange(n_studies + 1), all_labels)
    plt.axhline(y=pooled_pos - 0.5, color='grey', linestyle='--')  # Separator line
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.legend()
    
    if xlim:
        plt.xlim(xlim)
    
    if log_scale:
        plt.xscale('log')
        plt.grid(True, which="both", ls="-", alpha=0.3)
        # Add reference lines
        if "PLR" in title or "DOR" in title or "NLR" in title:
            plt.axvline(x=1, color='grey', linestyle='--', alpha=0.7)
    else:
        plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

# Sensitivity forest plot
studies = df['Study'].tolist()
sens_estimates = df['Sensitivity'].tolist()
sens_lower = df['Sens_CI_lower'].tolist()
sens_upper = df['Sens_CI_upper'].tolist()
create_forest_plot(studies, sens_estimates, sens_lower, sens_upper,
                   pooled_sens, sens_CI_lower, sens_CI_upper,
                   'Sensitivity Forest Plot', 'Sensitivity', (0, 1))

# Specificity forest plot
spec_estimates = df['Specificity'].tolist()
spec_lower = df['Spec_CI_lower'].tolist()
spec_upper = df['Spec_CI_upper'].tolist()
create_forest_plot(studies, spec_estimates, spec_lower, spec_upper,
                   pooled_spec, spec_CI_lower, spec_CI_upper,
                   'Specificity Forest Plot', 'Specificity', (0, 1))

# PLR forest plot
plr_estimates = df['PLR'].tolist()
plr_lower = df['PLR_CI_lower'].tolist()
plr_upper = df['PLR_CI_upper'].tolist()
create_forest_plot(studies, plr_estimates, plr_lower, plr_upper,
                   pooled_PLR, PLR_CI_lower, PLR_CI_upper,
                   'Positive Likelihood Ratio Forest Plot (Log Scale)', 
                   'PLR (log scale)', (0.1, 100), log_scale=True)

# NLR forest plot
nlr_estimates = df['NLR'].tolist()
nlr_lower = df['NLR_CI_lower'].tolist()
nlr_upper = df['NLR_CI_upper'].tolist()
create_forest_plot(studies, nlr_estimates, nlr_lower, nlr_upper,
                   pooled_NLR, NLR_CI_lower, NLR_CI_upper,
                   'Negative Likelihood Ratio Forest Plot (Log Scale)', 
                   'NLR (log scale)', (0.01, 10), log_scale=True)

# DOR forest plot with log scale
dor_estimates = df['DOR'].tolist()
dor_lower = df['DOR_CI_lower'].tolist()
dor_upper = df['DOR_CI_upper'].tolist()
create_forest_plot(studies, dor_estimates, dor_lower, dor_upper,
                   pooled_DOR, DOR_CI_lower, DOR_CI_upper,
                   'Diagnostic Odds Ratio Forest Plot (Log Scale)', 
                   'DOR (log scale)', (0.1, 100), log_scale=True)
```

```python
import numpy as np
import pandas as pd
from scipy.stats import chi2, norm
import matplotlib.pyplot as plt

# Data as DataFrame
data = pd.DataFrame({
    'Study': ['Yamamoto et al. 2022', 'Acampa et al. 2020', 'Giubbini et al. 2019',
              'Agostoni et al. 2018', 'Wells et al. 2017', 'Nkoulou et al. 2016'],
    'TP': [13, 12, 26, 5.5, 8.5, 7],
    'FP': [2, 3, 49, 0.5, 0.5, 4],
    'FN': [7, 2, 6, 1.5, 1.5, 3],
    'TN': [21, 8, 81, 24.5, 20.5, 14]
})

# Step 1: Calculate metrics
def calc_metrics(row):
    TP, FP, FN, TN = row['TP'], row['FP'], row['FN'], row['TN']
    sens = TP / (TP + FN) if (TP + FN) > 0 else np.nan
    spec = TN / (FP + TN) if (FP + TN) > 0 else np.nan
    plr = sens / (1 - spec) if (1 - spec) > 0 else np.nan
    nlr = (1 - sens) / spec if spec > 0 else np.nan
    dor = (TP * TN) / (FP * FN) if FP > 0 and FN > 0 else np.nan
    return pd.Series({'Sensitivity': sens, 'Specificity': spec, 'PLR': plr, 'NLR': nlr, 'DOR': dor})

metrics = data.apply(calc_metrics, axis=1)
df = pd.concat([data, metrics], axis=1)

# Step 2: Log-transform
df['lnPLR'] = np.log(df['PLR'])
df['lnNLR'] = np.log(df['NLR'])
df['lnDOR'] = np.log(df['DOR'])

# Step 3: Calculate SE for ln(DOR)
df['SE_lnDOR'] = np.sqrt(1/df['TP'] + 1/df['FP'] + 1/df['FN'] + 1/df['TN'])

# Step 4-9: Fixed-effect and random-effects weights for DOR
df['w_fixed'] = 1 / df['SE_lnDOR']**2
weighted_mean_lnDOR = np.sum(df['w_fixed'] * df['lnDOR']) / np.sum(df['w_fixed'])
Q = np.sum(df['w_fixed'] * (df['lnDOR'] - weighted_mean_lnDOR)**2)
S1 = np.sum(df['w_fixed'])
S2 = np.sum(df['w_fixed']**2)
k = df.shape[0]
tau2 = max(0, (Q - (k - 1)) / (S1 - (S2 / S1)))
df['w_random'] = 1 / (df['SE_lnDOR']**2 + tau2)

# Step 10: Pooled DOR with CI
pooled_lnDOR = np.sum(df['w_random'] * df['lnDOR']) / np.sum(df['w_random'])
pooled_DOR = np.exp(pooled_lnDOR)
SE_pooled_lnDOR = 1 / np.sqrt(np.sum(df['w_random']))
DOR_CI_lower = np.exp(pooled_lnDOR - norm.ppf(0.975) * SE_pooled_lnDOR)
DOR_CI_upper = np.exp(pooled_lnDOR + norm.ppf(0.975) * SE_pooled_lnDOR)

# Step 11: I² statistic
I2 = max(0, (Q - (k - 1)) / Q) * 100

# Step 12: Pooled Sensitivity/Specificity with CI
def logit(p):
    p = np.clip(p, 1e-10, 1 - 1e-10)
    return np.log(p / (1 - p))

def inv_logit(x):
    return 1 / (1 + np.exp(-x))

df['SE_sens'] = np.sqrt(df['Sensitivity'] * (1 - df['Sensitivity']) / (df['TP'] + df['FN']))
df['SE_spec'] = np.sqrt(df['Specificity'] * (1 - df['Specificity']) / (df['FP'] + df['TN']))
df['logit_sens'] = df['Sensitivity'].apply(logit)
df['logit_spec'] = df['Specificity'].apply(logit)

# Sensitivity calculations
df['w_sens_fixed'] = 1 / df['SE_sens']**2
Q_sens = np.sum(df['w_sens_fixed'] * (df['logit_sens'] - np.sum(df['w_sens_fixed'] * df['logit_sens']) / np.sum(df['w_sens_fixed']))**2)
S1_sens = np.sum(df['w_sens_fixed'])
S2_sens = np.sum(df['w_sens_fixed']**2)
tau2_sens = max(0, (Q_sens - (k - 1)) / (S1_sens - (S2_sens / S1_sens)))
df['w_sens_random'] = 1 / (df['SE_sens']**2 + tau2_sens)
pooled_logit_sens = np.sum(df['w_sens_random'] * df['logit_sens']) / np.sum(df['w_sens_random'])
pooled_sens = inv_logit(pooled_logit_sens)
SE_pooled_logit_sens = 1 / np.sqrt(np.sum(df['w_sens_random']))
sens_CI_lower = inv_logit(pooled_logit_sens - norm.ppf(0.975) * SE_pooled_logit_sens)
sens_CI_upper = inv_logit(pooled_logit_sens + norm.ppf(0.975) * SE_pooled_logit_sens)

# Specificity calculations
df['w_spec_fixed'] = 1 / df['SE_spec']**2
Q_spec = np.sum(df['w_spec_fixed'] * (df['logit_spec'] - np.sum(df['w_spec_fixed'] * df['logit_spec']) / np.sum(df['w_spec_fixed']))**2)
S1_spec = np.sum(df['w_spec_fixed'])
S2_spec = np.sum(df['w_spec_fixed']**2)
tau2_spec = max(0, (Q_spec - (k - 1)) / (S1_spec - (S2_spec / S1_spec)))
df['w_spec_random'] = 1 / (df['SE_spec']**2 + tau2_spec)
pooled_logit_spec = np.sum(df['w_spec_random'] * df['logit_spec']) / np.sum(df['w_spec_random'])
pooled_spec = inv_logit(pooled_logit_spec)
SE_pooled_logit_spec = 1 / np.sqrt(np.sum(df['w_spec_random']))
spec_CI_lower = inv_logit(pooled_logit_spec - norm.ppf(0.975) * SE_pooled_logit_spec)
spec_CI_upper = inv_logit(pooled_logit_spec + norm.ppf(0.975) * SE_pooled_logit_spec)

# Step 13: Pooled PLR and NLR with CI
pooled_PLR = pooled_sens / (1 - pooled_spec)
pooled_NLR = (1 - pooled_sens) / pooled_spec

# Delta method for PLR CI
var_sens = (pooled_sens * (1 - pooled_sens))**2 * (1 / np.sum(df['w_sens_random']))
var_spec = (pooled_spec * (1 - pooled_spec))**2 * (1 / np.sum(df['w_spec_random']))
var_logPLR = var_sens / (pooled_sens**2) + var_spec / ((1 - pooled_spec)**2)
SE_logPLR = np.sqrt(var_logPLR)
PLR_CI_lower = np.exp(np.log(pooled_PLR) - norm.ppf(0.975) * SE_logPLR)
PLR_CI_upper = np.exp(np.log(pooled_PLR) + norm.ppf(0.975) * SE_logPLR)

# Delta method for NLR CI
var_logNLR = var_sens / ((1 - pooled_sens)**2) + var_spec / (pooled_spec**2)
SE_logNLR = np.sqrt(var_logNLR)
NLR_CI_lower = np.exp(np.log(pooled_NLR) - norm.ppf(0.975) * SE_logNLR)
NLR_CI_upper = np.exp(np.log(pooled_NLR) + norm.ppf(0.975) * SE_logNLR)

# Step 14: Print results with CIs
print("=== Pooled Estimates with 95% Confidence Intervals ===")
print(f"Pooled Sensitivity: {pooled_sens:.3f} (95% CI: {sens_CI_lower:.3f}-{sens_CI_upper:.3f})")
print(f"Pooled Specificity: {pooled_spec:.3f} (95% CI: {spec_CI_lower:.3f}-{spec_CI_upper:.3f})")
print(f"Pooled PLR: {pooled_PLR:.3f} (95% CI: {PLR_CI_lower:.3f}-{PLR_CI_upper:.3f})")
print(f"Pooled NLR: {pooled_NLR:.3f} (95% CI: {NLR_CI_lower:.3f}-{NLR_CI_upper:.3f})")
print(f"Pooled DOR: {pooled_DOR:.3f} (95% CI: {DOR_CI_lower:.3f}-{DOR_CI_upper:.3f})")

# Calculate confidence intervals for all metrics - FIXED COLUMN NAMES
df['Sensitivity_CI_lower'] = np.maximum(0, df['Sensitivity'] - 1.96 * df['SE_sens'])
df['Sensitivity_CI_upper'] = np.minimum(1, df['Sensitivity'] + 1.96 * df['SE_sens'])
df['Specificity_CI_lower'] = np.maximum(0, df['Specificity'] - 1.96 * df['SE_spec'])
df['Specificity_CI_upper'] = np.minimum(1, df['Specificity'] + 1.96 * df['SE_spec'])
df['DOR_CI_lower'] = np.exp(df['lnDOR'] - 1.96 * df['SE_lnDOR'])
df['DOR_CI_upper'] = np.exp(df['lnDOR'] + 1.96 * df['SE_lnDOR'])

# Calculate SE for PLR and NLR for individual studies
df['SE_lnPLR'] = np.sqrt((1 - df['Sensitivity']) / (df['Sensitivity'] * (df['TP'] + df['FN'])) + 
                      df['Specificity'] / ((1 - df['Specificity']) * (df['FP'] + df['TN'])))
df['SE_lnNLR'] = np.sqrt(df['Sensitivity'] / ((1 - df['Sensitivity']) * (df['TP'] + df['FN'])) + 
                      (1 - df['Specificity']) / (df['Specificity'] * (df['FP'] + df['TN'])))

# Calculate CI for PLR and NLR for individual studies
df['PLR_CI_lower'] = np.exp(df['lnPLR'] - 1.96 * df['SE_lnPLR'])
df['PLR_CI_upper'] = np.exp(df['lnPLR'] + 1.96 * df['SE_lnPLR'])
df['NLR_CI_lower'] = np.exp(df['lnNLR'] - 1.96 * df['SE_lnNLR'])
df['NLR_CI_upper'] = np.exp(df['lnNLR'] + 1.96 * df['SE_lnNLR'])

# Create and display metric tables - FIXED COLUMN NAMES
def create_metric_table(metric, label, pooled_val, ci_lower, ci_upper):
    # Use consistent column names
    table = pd.DataFrame({
        'Study': df['Study'],
        f'{label}': df[metric],
        'Lower 95% CI': df[f'{metric}_CI_lower'],
        'Upper 95% CI': df[f'{metric}_CI_upper']
    })
    
    # Add pooled estimate as last row
    pooled_row = pd.DataFrame({
        'Study': ['Pooled Estimate'],
        f'{label}': [pooled_val],
        'Lower 95% CI': [ci_lower],
        'Upper 95% CI': [ci_upper]
    })
    return pd.concat([table, pooled_row], ignore_index=True)

# Generate all tables
sensitivity_table = create_metric_table(
    'Sensitivity', 'Sensitivity', pooled_sens, sens_CI_lower, sens_CI_upper
)
specificity_table = create_metric_table(
    'Specificity', 'Specificity', pooled_spec, spec_CI_lower, spec_CI_upper
)
plr_table = create_metric_table(
    'PLR', 'PLR', pooled_PLR, PLR_CI_lower, PLR_CI_upper
)
nlr_table = create_metric_table(
    'NLR', 'NLR', pooled_NLR, NLR_CI_lower, NLR_CI_upper
)
dor_table = create_metric_table(
    'DOR', 'DOR', pooled_DOR, DOR_CI_lower, DOR_CI_upper
)

# Print all tables
print("\n\n" + "="*70)
print("Sensitivity Results with 95% Confidence Intervals")
print("="*70)
print(sensitivity_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Specificity Results with 95% Confidence Intervals")
print("="*70)
print(specificity_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Positive Likelihood Ratio (PLR) Results")
print("="*70)
print(plr_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Negative Likelihood Ratio (NLR) Results")
print("="*70)
print(nlr_table.round(3))
print("\n" + "-"*70)

print("\n\n" + "="*70)
print("Diagnostic Odds Ratio (DOR) Results")
print("="*70)
print(dor_table.round(3))
print("\n" + "-"*70)

# Print heterogeneity metrics
print("\n\nHeterogeneity Metrics:")
print(f"I² Statistic: {I2:.1f}%")
print(f"Tau²: {tau2:.4f}")

# Step 15: Create forest plots
def create_forest_plot(studies, estimates, lower_bounds, upper_bounds, 
                       pooled_estimate, pooled_lower, pooled_upper,
                       title, xlabel, xlim=None, log_scale=False):
    plt.figure(figsize=(10, 6))
    n_studies = len(studies)
    y_pos = np.arange(n_studies)  # Positions for individual studies
    pooled_pos = n_studies  # Position for pooled estimate (below all studies)
    
    # Individual studies
    plt.errorbar(estimates, y_pos, 
                 xerr=[np.array(estimates) - np.array(lower_bounds), 
                       np.array(upper_bounds) - np.array(estimates)],
                 fmt='o', color='black', capsize=5, label='Individual Studies')
    
    # Pooled estimate
    plt.scatter(pooled_estimate, pooled_pos, color='red', s=100, zorder=3, 
                label='Pooled Estimate')
    plt.hlines(pooled_pos, pooled_lower, pooled_upper, color='red', linewidth=3)
    
    # Study labels
    all_labels = list(studies) + ['Pooled Estimate']
    plt.yticks(np.arange(n_studies + 1), all_labels)
    plt.axhline(y=pooled_pos - 0.5, color='grey', linestyle='--')  # Separator line
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.legend()
    
    if xlim:
        plt.xlim(xlim)
    
    if log_scale:
        plt.xscale('log')
        plt.grid(True, which="both", ls="-", alpha=0.3)
        # Add reference lines
        if "PLR" in title or "DOR" in title or "NLR" in title:
            plt.axvline(x=1, color='grey', linestyle='--', alpha=0.7)
    else:
        plt.grid(axis='x', linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.show()

# Sensitivity forest plot - USING CONSISTENT COLUMN NAMES
studies = df['Study'].tolist()
sens_estimates = df['Sensitivity'].tolist()
sens_lower = df['Sensitivity_CI_lower'].tolist()
sens_upper = df['Sensitivity_CI_upper'].tolist()
create_forest_plot(studies, sens_estimates, sens_lower, sens_upper,
                   pooled_sens, sens_CI_lower, sens_CI_upper,
                   'Sensitivity Forest Plot', 'Sensitivity', (0, 1))

# Specificity forest plot - USING CONSISTENT COLUMN NAMES
spec_estimates = df['Specificity'].tolist()
spec_lower = df['Specificity_CI_lower'].tolist()
spec_upper = df['Specificity_CI_upper'].tolist()
create_forest_plot(studies, spec_estimates, spec_lower, spec_upper,
                   pooled_spec, spec_CI_lower, spec_CI_upper,
                   'Specificity Forest Plot', 'Specificity', (0, 1))

# PLR forest plot
plr_estimates = df['PLR'].tolist()
plr_lower = df['PLR_CI_lower'].tolist()
plr_upper = df['PLR_CI_upper'].tolist()
create_forest_plot(studies, plr_estimates, plr_lower, plr_upper,
                   pooled_PLR, PLR_CI_lower, PLR_CI_upper,
                   'Positive Likelihood Ratio Forest Plot (Log Scale)', 
                   'PLR (log scale)', (0.1, 100), log_scale=True)

# NLR forest plot
nlr_estimates = df['NLR'].tolist()
nlr_lower = df['NLR_CI_lower'].tolist()
nlr_upper = df['NLR_CI_upper'].tolist()
create_forest_plot(studies, nlr_estimates, nlr_lower, nlr_upper,
                   pooled_NLR, NLR_CI_lower, NLR_CI_upper,
                   'Negative Likelihood Ratio Forest Plot (Log Scale)', 
                   'NLR (log scale)', (0.01, 10), log_scale=True)

# DOR forest plot with log scale
dor_estimates = df['DOR'].tolist()
dor_lower = df['DOR_CI_lower'].tolist()
dor_upper = df['DOR_CI_upper'].tolist()
create_forest_plot(studies, dor_estimates, dor_lower, dor_upper,
                   pooled_DOR, DOR_CI_lower, DOR_CI_upper,
                   'Diagnostic Odds Ratio Forest Plot (Log Scale)', 
                   'DOR (log scale)', (0.1, 100), log_scale=True)
```
